### Q.17) How can you replace all missing values in the 'Sales' column with the mean value of that column?

##### Note:- Let us assume there is null values in sales column

```python
df['Sales'].isnull().sum()
```

```python
mean_sales = df['Sales'].mean()
print("Mean of sales: ",mean_sales)

df['Sales'].fillna(mean_sales, inplace=True)
print(df["Sales"])
```
