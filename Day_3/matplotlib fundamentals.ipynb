## Single and Multiple plots in One Chart 

### Single Plot

```python
import matplotlib.pyplot as plt
plt.plot([1, 3, 2, 4], 'b-')
plt.show( )
```

```python
import matplotlib.pyplot as plt
x3 = np.arange(0.0, 6.0, 0.01) 
plt.plot(x3, [xi**2 for xi in x3], 'b-') 
plt.show()
```

### Multiple Plots


```python
import matplotlib.pyplot as plt
x4 = range(1, 5)
plt.plot(x4, [xi*1.5 for xi in x4])
plt.plot(x4, [xi*3 for xi in x4])
plt.plot(x4, [xi/3.0 for xi in x4])
plt.show()
```

## Plotting Different Curves 

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")

plt.legend()
plt.show()
```

## Matplotlib Style Formatting 

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

### Maplotlib Style format

```python
import matplotlib.pyplot as plt
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**2.7, 'g^', t, 2*t*t, 'ro', t, 3*t*t,"--")
plt.show()
```

## Labels and Legends

### Adding labels



```python
import matplotlib.pyplot as plt
plt.plot([1, 3, 2, 4])

plt.title('First Plot') # adding title

plt.xlabel('This is the X axis') #adding label
plt.ylabel('This is the Y axis') #adding Labels

plt.show()
```

### Adding a legend


```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['Normal','Fast','Slow']) #legend 
```

## Matplotlib Subplost

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(14,5))

x=  np.linspace(0, 10, 15)

ax1 = fig.add_subplot(2, 2, 1)
plt.plot(x, 2*x, '--')

ax2 = fig.add_subplot(2, 2, 2)
plt.plot(x, x**9, 'ro')

ax3 = fig.add_subplot(2, 2, 3)
plt.plot(x, x**3, 'g^')

ax4 = fig.add_subplot(2, 2, 4)
plt.plot(x, x**2, '-')
```

# Reading an external dataset

```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

## Line Plot


```python
import numpy as np
import matplotlib.pyplot as plt 

fig = plt.figure(figsize = (15,6))
ax = plt.axes()
ax.plot(df['math score'], 'b-'); 
```

## Scatter Plot


```python
import numpy as np
import matplotlib.pyplot as plt 

fig = plt.figure(figsize = (15,6))
ax = plt.axes()
ax.plot(df['math score'], df['reading score'] ,'ro'); 
```

## Histogram





```python
import numpy as np
import matplotlib.pyplot as plt 

fig = plt.figure(figsize = (15,6))
ax = plt.axes()


plt.hist(df['math score']); 
```

## Bar Chart

```python
import numpy as np
import matplotlib.pyplot as plt 

plt.bar(df['race/ethnicity'], df['math score'])

plt.show() 
```

## Horizontal Bar Chart


```python
import matplotlib.pyplot as plt 

plt.barh(df['race/ethnicity'], df['math score'])

plt.show() 
```

## Stacked Bar Chart


```python
df['gender']
df['reading score']
df['gender']
z2 = range(4)

plt.bar(df['gender'], df['math score'], color = 'b')
plt.bar(df['gender'],df['reading score'], color = 'r',
        bottom = df['reading score'])

plt.show()
```

The optional **bottom** parameter of the plt.bar() function allows us to specify a starting position for a bar. Instead of running from zero to a value, it will go from the bottom to value. The first call to plt.bar() plots the blue bars. The second call to plt.bar() plots the red bars, with the bottom of the red bars being at the top of the blue bars.

## Pie Chart

## Styles with Matplotlib Plots


```python
# View list of all available styles

print(plt.style.available)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-poster')
fig = plt.figure(figsize=(14,7))

x=  np.linspace(0, 10, 15)

plt.style.use('seaborn-deep')
ax1 = fig.add_subplot(3, 2, 1)

plt.style.use('seaborn-bright')
plt.plot(x, 2*x, '--')

plt.style.use('Solarize_Light2')
ax2 = fig.add_subplot(3, 2, 2)
plt.plot(x, x**9, 'ro')

plt.style.use('fivethirtyeight')
ax3 = fig.add_subplot(3, 2, 3)
plt.plot(x, x**3, 'g^')

plt.style.use('dark_background')
ax4 = fig.add_subplot(3, 2, 4)
plt.plot(x, x**2, '-')

plt.style.use('seaborn-ticks')
ax4 = fig.add_subplot(3, 2, 5)
plt.plot(x, x**2, '--')

plt.style.use('bmh')
ax4 = fig.add_subplot(3, 2, 6)
plt.plot(x, x**2, 'bs')
```
