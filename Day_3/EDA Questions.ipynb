### Q.2) Plot a barplot to show yearly profit?

```python
sns.set(style="darkgrid")

# Group the data by year and calculate the total profit for each year
yearly_profit = df.groupby('Year')['Profit'].sum().reset_index()

plt.figure(figsize=(8, 4))  
ax = sns.barplot(data=yearly_profit, x='Year', y='Profit')

plt.xlabel('Year')
plt.ylabel('Profit')
plt.title('Yearly Profit')

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

plt.show()
```

### Q.3) Plot CountPlot to show the frequency of Lead Column?

```python
sns.set(style="darkgrid")

plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='Lead') 

plt.xlabel('Lead')
plt.ylabel('Frequency')
plt.title('Frequency of Each Lead')

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

plt.show()
```

### Q.4) Plot the average profit for each month? 

```python
sns.set(style="darkgrid")

monthly_avg_profit = df.groupby('Month')['Profit'].mean().reset_index()

plt.figure(figsize=(8, 4))
ax = sns.barplot(data=monthly_avg_profit, x='Month', y='Profit')

plt.xlabel('Month')
plt.ylabel('Average Profit')
plt.title('Average Profit for Each Month')

plt.xticks(rotation=90)

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

plt.show()
```

### Q.5) Create a box plot to visualize the distribution & outliers of "Profit" column.


```python
plt.figure(figsize=(8, 6)) 
plt.boxplot(df['Profit'], flierprops={'marker': 'D', 'markerfacecolor': 'red', 'markersize': 6})

plt.xlabel('Profit')
plt.ylabel('Value')
plt.title('Distribution and Outliers of Profit')
plt.show()
```

### Q.6) Plot a pie chart of the product types column?

```python
product_counts = df['Product Type'].value_counts()

# Get the product types and their corresponding counts
product_types = product_counts.index.tolist()
product_counts = product_counts.tolist()

plt.figure(figsize=(8, 6))
plt.pie(product_counts, labels=product_types, autopct='%1.1f%%')
plt.title('Distribution of Product Types')
plt.legend(loc='best')
plt.show()
```

### Q.7) Create a bar chart of the total sales by state?

```python
state_sales = df.groupby('State')['Sale Price'].sum().reset_index()

state_sales = state_sales.sort_values('Sale Price', ascending=False)

plt.figure(figsize=(8, 4))  
bar_plot = plt.bar(state_sales['State'], state_sales['Sale Price'])

plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales by State')

for rect in bar_plot:
    height = rect.get_height()
    plt.annotate(f'{height:,.2f}', xy=(rect.get_x() + rect.get_width() / 2, height), xytext=(0, 3),
                 textcoords="offset points", ha='center', va='bottom')

plt.show()
```

### Q.8) How can you create a heatmap using Seaborn that shows the correlation between variables in a Pandas DataFrame?


```python
correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)

plt.title('Correlation Heatmap')
plt.show()
```

### Q.9) Load the iris dataset from Seaborn and create a line plot showing the changes in sepal length over time for all three species.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from Seaborn
iris = sns.load_dataset('iris')

# Create a line plot for sepal length for each species
plt.figure(figsize=(10, 6)) 
sns.lineplot(data=iris, x=iris.index, y='sepal_length', hue='species')


plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.title('Changes in Sepal Length for Iris Species')
plt.show()
```

### Q.10) Create a histogram of petal length for each species, with a different color for each.


```python
plt.figure(figsize=(8, 4)) 
sns.histplot(data=iris, x='petal_length', hue='species', element='step', stat='density', common_norm=False)


plt.xlabel('Petal Length')
plt.ylabel('Density')
plt.title('Histogram of Petal Length for Iris Species')
plt.show()
```

### Q.11)Create a scatter plot of sepal length vs. sepal width, with points colored according to species.

```python
plt.figure(figsize=(8, 4))  
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.show()
```

### Q.12)Create a box plot of sepal length for each species, with the box color set to a different color for each.


```python
plt.figure(figsize=(8, 4))  
sns.boxplot(data=iris, x='species', y='sepal_length', palette='Set3')

plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Box Plot of Sepal Length for Iris Species')
plt.show()
```

### Q.13)Load the tips dataset from Seaborn and create the distribution of total bills.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(8,4)) 
sns.histplot(data=tips, x='total_bill', kde=True)

plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bills')
plt.show()
```

### Q.14)Plot the relationship between total bill and tip amount

```python
plt.figure(figsize=(8, 4)) 
sns.scatterplot(data=tips, x='total_bill', y='tip')

plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title('Relationship between Total Bill and Tip Amount')
plt.show()
```

### Q.15) Plot the distribution of tip amounts compare between smokers and non-smokers?


```python
plt.figure(figsize=(8, 4))
sns.histplot(data=tips, x='tip', hue='smoker', element='step', stat='density', common_norm=False)

plt.xlabel('Tip Amount')
plt.ylabel('Density')
plt.title('Distribution of Tip Amounts for Smokers and Non-Smokers')
plt.show()
```

### Q.16) Create a bar chart showing the average total bill amount for each day of the week

```python
average_total_bill = tips.groupby('day')['total_bill'].mean().reset_index()

plt.figure(figsize=(8, 4))  
ax = sns.barplot(data=average_total_bill, x='day', y='total_bill')

plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill Amount')
plt.title('Average Total Bill Amount for Each Day of the Week')


for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

plt.show()
```

### Q.17) Plot a pie chart of the smoker column?

```python
smoker_counts = tips['smoker'].value_counts()

plt.figure(figsize=(8, 6)) 
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Smokers')
plt.legend(loc='upper right')
plt.show()
```
